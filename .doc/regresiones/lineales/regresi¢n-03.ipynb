```python
# AJUSTE POLINOMIAL - Versión manual
# En este ejemplo se muestra como realizar el ajuste polinomial de grado 1,
# de manera manual

# Se importan las librerías de trabajo
# ----------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt

# Datos de entrenamiento
# ----------------------------------------------------------------------------
X = np.array([6, 8, 10, 14, 18]).reshape(-1, 1)
y = [7, 9, 13, 17.5, 18]

# Generación del modelo gráfico
# ----------------------------------------------------------------------------
plt.figure()
plt.title('Precio de la Pizza contra el diámetro')
plt.xlabel('Diametro en pulgadas')
plt.ylabel('Precio en dólares')
plt.plot(X, y, 'b.', markersize=20)
plt.axis([0, 25, 0, 25])
plt.grid(True)

# Ajuste manual
# ----------------------------------------------------------------------------
plt.plot([0, 25], [1.97, 26.37], color='k', linestyle='-', linewidth=2)

plt.show()
```

```python
# VARIACIONES DE AJUSTE MANUAL POLINOMIAL

plt.figure()
plt.title('Precio de la Pizza contra el diámetro')
plt.xlabel('Diametro en pulgadas')
plt.ylabel('Precio en dólares')
plt.plot(X, y, 'b.', markersize=20)
plt.axis([0, 25, 0, 25])
plt.grid(True)

plt.plot([0, 25], [0, 28], color='k', linestyle='-', linewidth=2)
plt.plot([0, 25], [1, 29], color='k', linestyle='--', linewidth=2)
plt.plot([-1, 25], [0.5, 27], color='k', linestyle='-.', linewidth=2)
plt.show()
```
